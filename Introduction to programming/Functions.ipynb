### Funzioni

```python
# Define the function
def add_three(input_var):
    output_var = input_var + 3
    return output_var
```

Ogni funzione Ã¨ composta da due parti: un'intestazione e un corpo.

**header**

L'intestazione della funzione definisce il nome della funzione e i suoi argomenti.

Ogni intestazione di funzione inizia con def, che indica a Python che stiamo per definire una funzione.

Nell'esempio, il nome della funzione Ã¨ add_three.

Nell'esempio, l'argomento Ã¨ input_var. L'argomento Ã¨ il nome della variabile che verrÃ  utilizzata come input della funzione. Ã sempre racchiuso tra parentesi che compaiono subito dopo il nome della funzione. (Si noti che una funzione puÃ² anche non avere argomenti o avere piÃ¹ argomenti. Ne vedremo alcuni esempi piÃ¹ avanti nella lezione).

Per ogni funzione, le parentesi che racchiudono gli argomenti della funzione devono essere seguite dai due punti :.

**body**

Il corpo della funzione specifica il lavoro che la funzione svolge.

Ogni riga di codice nel corpo della funzione deve essere rientrata esattamente di quattro spazi. Per farlo, basta premere quattro volte la barra spaziatrice o premere una volta il tasto "Tab" sulla tastiera. (Con l'approfondimento di Python, potrebbe essere necessario rientrare il codice di piÃ¹ di quattro spazi, ma questo verrÃ  spiegato piÃ¹ avanti nel corso).

La funzione esegue il lavoro eseguendo tutte le righe rientrate dall'alto verso il basso.

Prende come input l'argomento, che nell'esempio Ã¨ input_var.

La funzione crea una nuova variabile output_var con il calcolo output_var = input_var + 3.

Quindi, l'ultima riga di codice, chiamata istruzione return, restituisce semplicemente il valore in output_var come output della funzione.

La cella di codice precedente definisce solo la funzione, ma non la esegue. I dettagli del corpo della funzione avranno piÃ¹ senso dopo la prossima cella di codice, quando la funzione verrÃ  effettivamente eseguita.

**Call a function**

Quando si esegue una funzione, si parla anche di "calling" della funzione.

Nella cella di codice sottostante, eseguiamo la funzione con 10 come valore di ingresso. Definiamo una nuova variabile new_number che viene impostata come uscita della funzione.

```python
# Run the function with 10 as input
new_number = add_three(10)

# Check that the value is 13, as expected
print(new_number)
```

In dettaglio,

- add_three(10) Ã¨ il valore che otteniamo come output quando forniamo 10 come valore di input_var e chiamiamo la funzione add_three(). Quando la funzione viene eseguita, esegue tutto il codice nel suo corpo, dall'inizio alla fine:
 - Per prima cosa calcola output_var = input_var + 3, che imposta output_var = 13.
 - L'ultima riga di codice Ã¨ l'istruzione return, che restituisce il valore di output_var, che Ã¨ 13.

Impostando nuovo_numero = add_three(10), si imposta nuovo_numero = 13.

Funzione add_three(), usiamo delle parentesi di chiusura vuote dopo il nome della funzione. Questo Ã¨ coerente con il modo in cui generalmente si scrivono le spiegazioni del codice Python e le parentesi vuote chiariscono che ci stiamo riferendo a una funzione, invece che a una variabile o a un altro oggetto Python. Queste parentesi dovrebbero essere sempre vuote, anche se la funzione ha degli argomenti.

Supponiamo che stiate aiutando un amico a calcolare la sua busta paga settimanale al netto delle tasse.

- L'amico si trova in una fascia fiscale del 12% (in altre parole, il 12% del suo stipendio viene tassato e lui ne porta a casa solo l'88%).
- Sono pagati a ore, con una tariffa di 15 dollari/ora.

La funzione seguente calcola la busta paga in base al numero di ore lavorate. La funzione Ã¨ piÃ¹ complicata rispetto al primo esempio, perchÃ© contiene piÃ¹ righe di codice e commenti. Come nell'esempio precedente, la funzione ha un unico argomento (num ore). Nel corpo della funzione, noi:

 - Usare il valore di num_hours per specificare il valore di una nuova variabile pay_pretax.
 - Utilizzare il valore di pay_pretax per specificare il valore di una nuova variabile pay_aftertax.
 - Restituire il valore della variabile pay_aftertax.

```python
def get_pay(num_hours):
    # Pre-tax pay, based on receiving $15/hour
    pay_pretax = num_hours * 15
    # After-tax pay, based on being in 12% tax bracket
    pay_aftertax = pay_pretax * (1 - .12)
    return pay_aftertax
```

```python
# Calculate pay based on working 40 hours
pay_fulltime = get_pay(40)
print(pay_fulltime)
```

Chiamiamo questa funzione nello stesso modo in cui abbiamo chiamato la prima funzione. La cella di codice successiva calcola lo stipendio, basato su 40 ore di lavoro. (Al netto delle tasse, Ã¨ di 528 dollari).

```python
pay_parttime = get_pay(32)
print(pay_parttime)
```

Avendo scritto una funzione, Ã¨ possibile calcolare la retribuzione per diverse ore senza dover riscrivere tutto il codice dei calcoli.

Le funzioni possono aiutare a evitare errori nel codice e a risparmiare molto tempo. In generale, quando si scrive, si dovrebbe puntare a scrivere il meno possibile, perchÃ© ogni volta che si digita un calcolo Ã¨ un'altra occasione per introdurre accidentalmente un refuso o un errore.

Le variabili definite all'interno del corpo della funzione non possono essere accessibili al di fuori della funzione. Ad esempio, la cella di codice successiva presenta un errore, perchÃ© pay_aftertax esiste solo all'interno della funzione.
Si otterrÃ  lo stesso errore se si cerca di stampare pay_pretax o num_hours. Per questo motivo, se avete bisogno di informazioni da una funzione, dovete assicurarvi che queste appaiano nell'istruzione return alla fine della funzione.

Per 'scope' di una variabile si intende la parte del codice in cui Ã¨ accessibile. Le variabili definite all'interno di una funzione (come pay_aftertax) hanno un ambito locale solo di quella funzione. Tuttavia, come abbiamo visto, le variabili definite al di fuori di tutte le funzioni (come pay_parttime) hanno un ambito globale e possono essere accessibili ovunque.

**PiÃ¹ variabili**

Per definire una funzione con piÃ¹ argomenti, Ã¨ sufficiente aggiungere altri argomenti all'interno delle parentesi nel titolo della funzione e separarli con una virgola.

Lo facciamo con la funzione get_pay_with_more_inputs(), che calcola una busta paga settimanale in base a tre argomenti:

- num_hours = numero di ore lavorate in una settimana
- hourly_wage = il salario orario (in dollari/ora)
- tax_bracket = percentuale del salario che viene tolta per le tasse

```python
def get_pay_with_more_inputs(num_hours, hourly_wage, tax_bracket):
    # Pre-tax pay
    pay_pretax = num_hours * hourly_wage
    # After-tax pay
    pay_aftertax = pay_pretax * (1 - tax_bracket)
    return pay_aftertax
```

```python
higher_pay_aftertax = get_pay_with_more_inputs(40, 24, .22)
print(higher_pay_aftertax)
```

Quindi, per chiamare la funzione, Ã¨ necessario fornire un valore per ciascun input, sempre separato da una virgola.

Nella cella di codice qui sotto, calcoliamo la retribuzione al netto delle tasse per una persona che lavora 40 ore, guadagna 24 dollari l'ora e si trova in una fascia fiscale del 22%.

```python
same_pay_fulltime = get_pay_with_more_inputs(40, 15, .12)
print(same_pay_fulltime)
```

A seconda di come si intende utilizzare questa nuova funzione get_pay_with_more_inputs(), puÃ² essere piÃ¹ utile della funzione originale get_pay(), perchÃ© affronta piÃ¹ casi. Invece di assumere potenzialmente in modo errato la retribuzione oraria e la fascia fiscale, la nuova funzione consente all'utente di specificare i valori corretti. Tuttavia, se si Ã¨ certi che il salario orario e la fascia fiscale non dovranno cambiare, la nuova funzione Ã¨ piÃ¹ complicata del necessario. In generale, quando si definiscono le funzioni, Ã¨ necessario considerare quanta flessibilitÃ  Ã¨ necessaria, in base al caso d'uso.
