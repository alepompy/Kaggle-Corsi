### Plottare i dati
Ora che il set di dati è caricato nel notebook, è necessaria solo una riga di codice per creare un grafico a linee!

````python
# Line chart showing daily global streams of each song 
sns.lineplot(data=spotify_data)
````

Come puoi vedere sopra, la riga di codice è relativamente breve e ha due componenti principali:

- `sns.lineplot` dice al notebook che vogliamo creare un grafico a linee.
    - Ogni comando che impari in questo corso inizierà con sns, che indica che il comando proviene dal pacchetto seaborn. Ad esempio, utilizziamo sns.lineplot per creare grafici a linee. Presto imparerai che utilizziamo sns.barplot e sns.heatmap rispettivamente per creare grafici a barre e mappe di calore.
- `data=spotify_data` seleziona i dati che verranno utilizzati per creare il grafico.
Tieni presente che utilizzerai sempre lo stesso formato quando crei un grafico a linee e l'unica cosa che cambia con un nuovo set di dati è il nome del set di dati. Quindi, se stai lavorando con un set di dati diverso chiamato financial_data, ad esempio, la riga di codice apparirà come segue:

                                                    `sns.lineplot(dati=dati_finanziari)`   
A volte ci sono dettagli aggiuntivi che vorremmo modificare, come la dimensione della figura e il titolo del grafico. Ognuna di queste opzioni può essere facilmente impostata con una singola riga di codice.

````python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# Line chart showing daily global streams of each song 
sns.lineplot(data=spotify_data)
````

La prima riga di codice imposta la dimensione della figura a 14 pollici (in larghezza) per 6 pollici (in altezza). Per impostare la dimensione di qualsiasi figura, è sufficiente copiare la stessa riga di codice che appare. Quindi, se desideri utilizzare una dimensione personalizzata, modifica i valori forniti di 14 e 6 con la larghezza e l'altezza desiderate.

La seconda riga di codice imposta il titolo della figura. Si noti che il titolo deve essere sempre racchiuso tra virgolette (**"..."**)!

### Plotta un sottoinsieme di dati
Finora, hai imparato come tracciare una linea per ogni colonna nel set di dati. In questa sezione imparerai come tracciare un sottoinsieme delle colonne.

Inizieremo stampando i nomi di tutte le colonne. Questo viene fatto con una riga di codice e può essere adattato a qualsiasi set di dati semplicemente sostituendo il nome del set di dati (in questo caso, spotify_data).

````python
list(spotify_data.columns)
````

Nella successiva cella di codice, tracciamo le linee corrispondenti alle prime due colonne del set di dati.

````python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# Line chart showing daily global streams of 'Shape of You'
sns.lineplot(data=spotify_data['Shape of You'], label="Shape of You")

# Line chart showing daily global streams of 'Despacito'
sns.lineplot(data=spotify_data['Despacito'], label="Despacito")

# Add label for horizontal axis
plt.xlabel("Date")
````

Le prime due righe di codice impostano il titolo e la dimensione della figura (e dovrebbero sembrare molto familiari!).

Le due linee successive aggiungono ciascuna una linea al grafico a linee. Ad esempio, considera il primo, che aggiunge la riga per "Shape of You":

````python
# Line chart showing daily global streams of 'Shape of You'
sns.lineplot(data=spotify_data['Shape of You'], label="Shape of You")
````

Questa riga è molto simile al codice che abbiamo usato quando abbiamo tracciato ogni riga nel set di dati, ma presenta alcune differenze chiave:

Invece di impostare data=spotify_data, impostiamo data=spotify_data['Shape of You']. In generale, per tracciare una sola colonna, utilizziamo questo formato mettendo il nome della colonna tra virgolette singole e racchiudendolo tra parentesi quadre. (Per assicurarti di specificare correttamente il nome della colonna, puoi stampare l'elenco di tutti i nomi delle colonne usando il comando che hai imparato sopra.)
Aggiungiamo anche label="Shape of You" per far apparire la linea nella legenda e impostare l'etichetta corrispondente.
L'ultima riga di codice modifica l'etichetta per l'asse orizzontale (o asse x), dove l'etichetta desiderata è racchiusa tra virgolette (**"..."**).
