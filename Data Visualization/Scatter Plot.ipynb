## Scatter Plot
In questo tutorial imparerai come creare grafici a dispersione avanzati.

Come sempre, iniziamo configurando l'ambiente di codifica. (Questo codice Ã¨ nascosto, ma puoi scoprirlo facendo clic sul pulsante "Codice" immediatamente sotto questo testo, a destra.)

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

Lavoreremo con un set di dati (sintetico) di spese assicurative, per vedere se riusciamo a capire perchÃ© alcuni clienti pagano piÃ¹ di altri.


```python
# Path of the file to read
insurance_filepath = "D:/Users/Alessio/OneDrive/Python/Kaggle/Data Visualization/insurance.csv"

# Read the file into a variable insurance_data
insurance_data = pd.read_csv(insurance_filepath)
```

```python
insurance_data.head()
```

### Regressione Lineare
Per creare un semplice **grafico a dispersione**, utilizziamo il comando sns.scatterplot e specifichiamo i valori per:

- l'asse x orizzontale (x=insurance_data['bmi']), e
- l'asse verticale delle y (y=insurance_data['addebiti']).

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

Il grafico a dispersione sopra suggerisce che l'indice di massa corporea (BMI) e gli oneri assicurativi sono **correlati positivamente**, dove i clienti con un BMI piÃ¹ elevato in genere tendono anche a pagare di piÃ¹ i costi assicurativi. (Questo schema ha senso, poichÃ© un BMI elevato Ã¨ in genere associato a un rischio piÃ¹ elevato di malattie croniche.)

Per ricontrollare la forza di questa relazione, potresti voler aggiungere una **linea di regressione** o la linea che meglio si adatta ai dati. Lo facciamo cambiando il comando in sns.regplot

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

### Grafici a dispersione con codice colore
Possiamo usare i grafici a dispersione per visualizzare le relazioni tra (non due, ma...) tre variabili! Un modo per farlo Ã¨ codificare a colori i punti.

Ad esempio, per capire come il fumo influisca sulla relazione tra BMI e costi assicurativi, possiamo codificare a colori i punti per "smoker" e tracciare le altre due colonne ("bmi", "charges") sugli assi.

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

Questo grafico a dispersione mostra che mentre i non fumatori tendono a pagare leggermente di piÃ¹ con l'aumento del BMI, i fumatori pagano MOLTO di piÃ¹.

Per enfatizzare ulteriormente questo fatto, possiamo usare il comando `sns.lplot` per aggiungere due linee di regressione, corrispondenti a fumatori e non fumatori. (Noterai che la linea di regressione per i fumatori ha una pendenza molto piÃ¹ ripida, rispetto alla linea per i non fumatori!)

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

Il comando sns.lplot sopra funziona in modo leggermente diverso rispetto ai comandi che hai imparato finora:

Invece di impostare `x=insurance_data['bmi']` per selezionare la colonna 'bmi' in Insurance_data, impostiamo `x="bmi"` per specificare solo il nome della colonna.  
Allo stesso modo, `y="charges"` e `hue="smoker"` contengono anche i nomi delle colonne.   
Specifichiamo il set di dati con `data=insurance_data`.  
Infine, c'Ã¨ un'altra trama che imparerai, che potrebbe sembrare leggermente diversa da come sei abituato a vedere le trame a dispersione. Di solito, utilizziamo grafici a dispersione per evidenziare la relazione tra due variabili continue (come "bmi" e "charges"). Tuttavia, possiamo adattare il design del grafico a dispersione per presentare una variabile categoriale (come "smoker") su uno degli assi principali. Ci riferiremo a questo tipo di grafico come a un **grafico a dispersione categoriale** e lo creeremo con il comando `sns.swarmplot`.

```python
sns.swarmplot(x=insurance_data['smoker'],
              y=insurance_data['charges'])
```

Tra le altre cose, questa trama ci mostra che:

- in media, i non fumatori pagano meno dei fumatori e
- i clienti che pagano di piÃ¹ sono i fumatori; mentre i clienti che pagano di meno sono non fumatori.
