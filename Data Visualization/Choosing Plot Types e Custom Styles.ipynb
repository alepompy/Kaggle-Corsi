## Choosing Plot Types e Custom Styles
In questo corso hai imparato a creare diversi tipi di grafici. Ora organizzerai le tue conoscenze, prima di apprendere alcuni comandi rapidi che puoi utilizzare per cambiare lo stile dei tuoi grafici.



Dal momento che non  sempre facile decidere come raccontare al meglio la storia dietro i tuoi dati, abbiamo suddiviso i tipi di grafico in tre grandi categorie per aiutarti in questo.

1. **Tendenze** - Una tendenza  definita come un modello di cambiamento.
    - sns.lineplot - I **grafici a linee** sono i migliori per mostrare le tendenze in un periodo di tempo e pi linee possono essere utilizzate per mostrare le tendenze in pi di un gruppo.
1. **Relazione**: esistono diversi tipi di grafici che puoi utilizzare per comprendere le relazioni tra le variabili nei dati.
    - sns.barplot - I **grafici a barre** sono utili per confrontare quantit corrispondenti a diversi gruppi.
    - sns.heatmap - Le **mappe di calore** possono essere utilizzate per trovare modelli codificati a colori nelle tabelle di numeri.
    - sns.scatterplot - I **grafici a dispersione** mostrano la relazione tra due variabili continue; se codificata a colori, possiamo anche mostrare la relazione con una terza variabile categoriale.
    - sns.regplot - L'inclusione di una **linea di regressione** nel grafico a dispersione semplifica la visualizzazione di qualsiasi relazione lineare tra due variabili.
    - sns.lmplot - Questo comando  utile per disegnare **pi linee di regressione**, se il grafico a dispersione contiene pi gruppi con codice colore.
    - sns.swarmplot - I **grafici a dispersione categoriali** mostrano la relazione tra una variabile continua e una variabile categoriale.
1. **Distribuzione**: visualizziamo le distribuzioni per mostrare i possibili valori che possiamo aspettarci di vedere in una variabile, insieme alla loro probabilit.
    - sns.histplot - Gli **istogrammi** mostrano la distribuzione di una singola variabile numerica.
    - sns.kdeplot - I **grafici KDE** (o KDE 2D) mostrano una distribuzione uniforme e stimata di una singola variabile numerica (o due variabili numeriche).
    - sns.jointplot - Questo comando  utile per visualizzare **simultaneamente un grafico 2D di KDE** con i corrispondenti grafici di KDE per ogni singola variabile.

### Cambiare stile con Seaborn
Tutti i comandi hanno fornito uno stile piacevole e predefinito a ciascuno dei grafici. Tuttavia, potresti trovare utile personalizzare l'aspetto delle tue trame e, per fortuna, questo pu essere ottenuto semplicemente aggiungendo un'altra riga di codice!

Come sempre, dobbiamo iniziare configurando l'ambiente di codifica. (Questo codice  nascosto, ma puoi scoprirlo facendo clic sul pulsante "Codice" immediatamente sotto questo testo, a destra.)

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

Lavoreremo con lo stesso codice utilizzato per creare un grafico a linee in un tutorial precedente. Il codice seguente carica il set di dati e crea il grafico.

```python
# Path of the file to read
spotify_filepath = "D:/Users/Alessio/OneDrive/Python/Kaggle/Data Visualization/spotify.csv"

# Read the file into a variable spotify_data
spotify_data = pd.read_csv(spotify_filepath, index_col="Date", parse_dates=True)

# Line chart 
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
```

Possiamo cambiare rapidamente lo stile della figura in un tema diverso con una sola riga di codice.

```python
# Change the style of the figure to the "dark" theme
sns.set_style("dark")

# Line chart 
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
```
